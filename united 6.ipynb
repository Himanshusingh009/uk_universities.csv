```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
df.dtypes
```

```python
df['Pclass'].value_counts()
```

```python
categorical_features = ['Sex','Embarked','Pclass']
```

```python
columns_transformer = ColumnTransformer(
    transformers=[
        ('encoder',OneHotEncoder(drop='first'),categorical_features)
    ],
    remainder='passthrough' # keep other cloumns
)
```

```python
encoded_array = columns_transformer.fit_transform(df)
```

```python
encoded_df = pd.DataFrame(encoded_array.toarray() if hasattr(encoded_array,'toarray') else encoded_array)
```
