# Displots
Displot stands for distribution plot, it takes as input an array and plots a curve corresponding to the distribution of points in the array.

```python
# Import Seaborn
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
sns.displot([0, 1, 2, 3, 4, 5])

plt.show()
```

```python
# Plotting a Displot Without the Histogram
sns.displot([0, 1, 2, 3, 4, 5], kind="kde")

plt.show()
```

```python
sns.set(style="white")

# Generate a random univariate dataset
rs = np.random.RandomState(10)
d = rs.normal(size=100)

# Plot a histogram with multiple colors
sns.distplot(d, kde=True, hist=True, bins=10)
plt.show()
```

# Histplot
Seaborn Histplot is used to visualize the univariate set of distributions(single variable). It plots a histogram, with some other variations like kdeplot and rugplot.

```python
sns.histplot([0, 1, 2, 3, 4, 5], kde=True, color="m")
plt.show()
```

```python
sns.set(style="white")

# Generate a random univariate dataset
rs = np.random.RandomState(10)
d = rs.normal(size=100)

# Plot a simple histogram and kde
sns.histplot(d, kde=True, color="m")
plt.show()
```

# Lineplot
The line plot is one of the most basic plots in the seaborn library. This plot is mainly used to visualize the data in the form of some time series, i.e. in a continuous manner.

```python
sns.set(style="dark")
fmri = sns.load_dataset("fmri")

# events and regions
sns.lineplot(x="timepoint",# waqt ke saath signal change ka pattern
             y="signal",# brain signal ki intensity
             hue="region",# alag-alag colors brain ke regions ke liye
             style="event",# alag-alag line styles jaise solid, dashed (stimulus ya rest events ke liye)
             data=fmri)
plt.show()
```

# countplot

```python
# read a tips.csv file from seaborn library
df = sns.load_dataset('tips')

# count plot on single categorical variable
sns.countplot(x ='sex', data = df)

plt.show()
```

```python
# read a tips.csv file from seaborn library
df = sns.load_dataset('tips')

# count plot on two categorical variable
sns.countplot(x ='sex', hue = "smoker", data = df)

plt.show()
```

```python
# read a tips.csv file from seaborn library
df = sns.load_dataset('tips')

# count plot along y axis
sns.countplot(y ='sex', hue = "smoker", data = df)

plt.show()
```

```python
# read a tips.csv file from seaborn library
df = sns.load_dataset('tips')

# use a different colour palette in count plot
sns.countplot(x ='sex', data = df, palette = "Set2")

plt.show()
```

```python
# Load the Titanic dataset from seaborn library
df = sns.load_dataset('titanic')

sns.countplot(x='class', hue='sex', data=df, color="salmon")

plt.show()
```

# heatmap

```python
# Generating a 10x10 matrix of random numbers
data = np.random.randint(1, 100, (10, 10))

sns.heatmap(data)
plt.show()
```

```python
data = np.random.randint(1, 100, (10, 10))

sns.heatmap(data, vmin=30, vmax=70)
plt.show()
```

```python
data = np.random.randint(1, 100, (10, 10))

sns.heatmap(data, cmap='tab20')
plt.show()
```

```python
data = np.random.randint(1, 100, (10, 10))

sns.heatmap(data, cmap='coolwarm', center=50)
plt.show()
```
