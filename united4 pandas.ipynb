## Pandas Introduction
- Pandas is a Python library used for working with data sets.

- It has functions for analyzing, cleaning, exploring, and manipulating data.

- The name "Pandas" has a reference to both "Panel Data", and "Python Data Analysis" and was created by Wes McKinney in 2008.

### Why Use Pandas?
- Pandas allows us to analyze big data and make conclusions based on statistical theories.
- Pandas can clean messy data sets, and make them readable and relevant.
- Relevant data is very important in data science.
### What Can Pandas Do?
Pandas gives you answers about the data. Like:

- Is there a correlation between two or more columns?
- What is average value?
- Max value?
- Min value?
Pandas are also able to delete rows that are not relevant, or contains wrong values, like empty or NULL values. This is called cleaning the data.

```python
# Create a simple Pandas Series from a list
import pandas as pd

a = [1, 7, 2]

myvar = pd.Series(a)

print(myvar)
```

```python
# Return the first value of the Series
print(myvar[1])
```

```python
# Create your own labels
a = [1, 7, 2]

myvar = pd.Series(a, index = ["x", "y", "z"])

print(myvar)
```

```python
# Create a simple Pandas Series from a dictionary
calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories) 

print(myvar)
```

```python
# Create a Series using only data from "day1" and "day2"
calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories, index = ["day1", "day2"])

print(myvar)
```

### DataFrames
Data sets in Pandas are usually multi-dimensional tables, called DataFrames.

Series is like a column, a DataFrame is the whole table.

```python
# Create a DataFrame from two Series
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}
#load data into a DataFrame object:
myvar = pd.DataFrame(data)

print(myvar)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Pie chart for Passed vs Failed
passed_counts = df['Passed'].value_counts()

# plt.figure(figsize=(6, 6))

plt.pie(passed_counts, labels=passed_counts.index, autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
plt.title('Pass vs Fail Distribution')
plt.axis('equal')
plt.show()
```

```python
# Add a column for average
df['Average'] = df[['Math', 'Science', 'English']].mean(axis=1)

plt.figure(figsize=(10, 5))
plt.plot(df['Name'], df['Average'], marker='o', color='green')
plt.title('Average Score per Student')
plt.xlabel('Student Name')
plt.ylabel('Average Score')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.bar(df['Name'], df['Math'], color='skyblue')
plt.title('Math Scores of Students')
plt.xlabel('Student Name')
plt.ylabel('Math Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.scatter(df['Math'], df['Science'], color='orange')
plt.title('Math vs Science Scores')
plt.xlabel('Math Score')
plt.ylabel('Science Score')
plt.grid(True)
plt.show()
```

## YData Profiling (formerly Pandas Profiling)
YData Profiling is an open-source Python library used for automated Exploratory Data Analysis (EDA). It generates an interactive HTML report from a pandas DataFrame with just a single line of code.
- Overview of missing values
- Descriptive statistics (mean, std, min, max, etc.)
- Correlations (Pearson, Spearman, Kendall)
- Histograms, distributions
- Duplicate row detection
- Highly correlated features
- Warnings about suspicious data patterns

```python
# pip install ydata-profiling
```

```python
import pandas as pd
from ydata_profiling import ProfileReport
```

```python
# Load data
df = pd.read_csv('data.csv')
```

```python
# Create profile report
profile = ProfileReport(df, title="My Data Report", explorative=True)
```

```python
# Show report in a Jupyter Notebook
profile.to_notebook_iframe()
```

```python
# Or save as an HTML file
profile.to_file("my_data_profile.html")
```
